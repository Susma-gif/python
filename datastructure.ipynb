```python
dicitionary is mutable data type 
you can add, update and delete in dictionary
```

```python
details={
    'name':['Manisha','Arya'],
    'Age':[22,12],
    'address':['Nakku','nepal']}
details 
```

```python
# key
for key in details.keys():
    print(key)
```

```python
# Value 
for value in details.values():
    print(value)
```

```python
#items
for key in details.items():
    print(f"{key}:{value}")
```
