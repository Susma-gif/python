# CSV (comma seperated value)

# Creating file in CSV

```python
import csv
```

```python
countries_destination = [
    ['Countries', 'Destination', 'Price'],
    ['USA', 'Newyork', '$45000'],
    ['Italy', 'Rome', '$25000'],
    ['Nepal', 'dhorpatan', '$12000']
]
```

```python
#open(file_name, operation,newline) 3 arguments goes in creating CSV file
with open('countries_destination.csv','w',newline='') as file:
    writer=csv.writer(file)
    writer.writerows(countries_destination)
```

# reading a CSV file

```python
with open('countries_destination.csv','r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

# Dicitionary Data with creating file
# for CSV and dictionary we import csv

```python
personal_details=[
    { "Name":"Manisha" ,
      "Age":22 ,
      "Address":"Nakkhu"
    },
    { "Name":"Arya" ,
      "Age":12,
      "Address":"Nepal"
    }, 
    { "Name":"Ramiro" ,
      "Age":35 ,
      "Address":"bhaise" }
]
```

# Write dictionary value in CSV file

```python
with open('personal_details.csv','w',newline='') as file:
    columns=['Name','Age','Address']
    writer=csv.DictWriter(file,fieldnames=columns)
    writer.writeheader()
    writer.writerows(personal_details)
```

```python
import csv

personal_details = [
    {"Name": "Inges", "Age": 22, "Address": "Nakkhu"},
    {"Name": "Sonia", "Age": 12, "Address": "Nepal"},
    {"Name": "Ramiro", "Age": 35, "Address": "bhaise"}
]

with open('personal_details.csv', 'w', newline='') as file:
    columns = ['Name', 'Age', 'Address']
    writer = csv.DictWriter(file, fieldnames=columns)
    writer.writeheader()
    writer.writerows(personal_details)
```

# read dictionary file

```python
with open('personal_details.csv','r') as file:
    reader=csv.DictReader(file)
    for row in reader:
        print(row)
```

# JSON file
# for json import json

```python
# dictionary and Json has same format
# write ->dump
# read-> load 
```

# write file in JSON

```python
import json

personal_details = [
    {"Name": "Inges", "Age": 22, "Address": "Nakkhu"},
    {"Name": "Sonia", "Age": 12, "Address": "Nepal"},
    {"Name": "Ramiro", "Age": 35, "Address": "bhaise"}
]
```

```python
with open('personal_details.json','w') as file:
    json.dump(personal_details,file)
```

# Read file in Json

```python
with open('personal_details.json','r') as file:
    data=json.load(file)
    for row in data:
        print(row)
```
