```python
def calculate_bonus(salary, years_of_service):
    print("Salary:", salary)
    print("Service:", years_of_service)

    if years_of_service >= 10:
        bonus = 0.15 * salary
        print("Bonus 15%")
    elif years_of_service >= 5:
        bonus = 0.10 * salary
        print("Bonus 10%")
    else:
        bonus = 0.05 * salary
        print("Bonus 5%")

    return bonus
```

```python
calculate_bonus(500000,5)
```

```python
final_bonus=calculate_bonus(50000,7)
print("final bonus:",final_bonus)
```
