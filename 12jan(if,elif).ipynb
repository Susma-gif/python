# matchcase

```python
choice=int(input("Enter the number from 1-5:"))
match choice:
    case 1:
        print("you have ordered lunch")
    case 2:
        print("You have ordered Dinner")
    case 3:
        print("You have ordered Breakfast")
    case 4:
        print("you have ordered snacks")
    case 5:
        print("Thankyou for choosing us")
    case _:
        print("Invalid value")
```

```python
a=int(input("Enter the first number:"))
b =int(input("Enter the second number:"))
choice=input("Enter value among +,-,*,/:")
#using match case to meet all conditions
match choice:
    case'+':
        add=a+b
        print("Addition of a and b is :",add)
    case'-':
        sub=a-b
        print("Subtraction of a and b is:",sub)
    case'*':
        mul=a*b
        print("Multiplication of and b is:",mul)
    case'/':
        div=a/b
        print("Division of a and b is:",div)
    case _:
        print("Invalid Value")

        
```

# Break and continue

```python
#Condition True:
    break (program stop)
# condition True:
    continue (value skip)
```

```python
#Find numbers from 1 to 10 but program needs to stopped when it finds number 5
for i in range(1,11):
    if i==5:
        break
    print(i)
```

```python
#Find numbers from 1 to 10 but program needs to continued when it finds number 5
for i in range(1,11):
    if i==5:#skips 5 and prints rest
        continue
    print(i)
```

```python
#user input 5 times,if user inputs 0 value then program needs to be stopped
for i in range(5):
    n=int(input("Enter the number:"))
    if n==0:
        print("You have entered 0.Programme stopped")
        break
```
