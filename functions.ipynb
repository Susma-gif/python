# Default Argument and Keyword Argument

# Default argument

```python
def person(name="unknown",age=0,address="unknown"):
    print("Name:",name)
    print("age:",age)
    print("address:",address)
```

```python
person("Manisha",22,"Nakkhu")
```

```python
person(name="ABC",age=22,address="Nakkhu")# we can also pass argument in default argument
```

```python
def person(name="Arya",age=0,address="Nepal"):
    print("Name:",name)
    print("Age:",age)
    print("address:",address)
```

```python
person() # if we didnot pass any argument inside the bracket, then it will give default value
```

# (All Arguments)*Args, (All keyword Arguments)**kwargs

```python
def phone(phone_name,*args):
    print("phone name:",phone_name)
```

```python
phone("Nothing")
```

```python
phone("S22",2022)
```

# Adding loop in Argument

```python
def phone(phone_name,*args):
    print("phone name:",phone_name)
    for each_arg in args:
        print(each_arg) # we can use either for loop or all arguments code
    print("All arguments:",args)
```

```python
phone("Nothing","s22",2025)
```

```python
def phone_details(phone_name,**kwargs):
    print("Phone Name:",phone_name)
    print(kwargs)
```

```python
phone_details("Nothing",model="s25",year=2025, os="Android 14")
```

```python
def phone_details1(phone_name,**kwargs):
    print("Phone Name:",phone_name)
    print(kwargs)
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
phone_details1("Nothing",model="s25",year=2025,os="Android14")
```
