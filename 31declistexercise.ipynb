```python
#find the minimum and maximum age
ages=[19,23,19,25,21,20,25,26,25,24]
ages.sort()
min_age=ages[0]
max_age=ages[-1]
print(f"Sorted Ages:{ages}")
print(f"Min Age:{min_age},Max age:{max_age}")
```

```python
#Add min amd max age again to the list
ages=[19,23,19,25,21,20,25,26,25,24]
ages.sort()
min_age=ages[0]
max_age=ages[-1]
ages.append(min_age)
ages.append(max_age)
print(f"List after adding min/max:{ages}")
```

```python
#Calculate the median age(middle value or average of 2 middle value)
ages=[19,23,19,25,21,20,25,26,25,24]
len(ages)
```

```python
median=(ages[len(ages)//2])
print(f"Median Age:{median}")
```

```python
#find the average age
ages=[19,23,19,25,21,20,25,26,25,24]
average_age=sum(ages)/len(ages)
print(average_age)
```

```python
#calculate the ranges(max-min)
ages=[19,23,19,25,21,20,25,26,25,24]
age_range=max(ages)-min(ages)
print(age_range)
```

```python
#use the abs function to compare abs(min-average) vs abs(max-average)
ages=[19,23,19,25,21,20,25,26,25,24]
average=sum(ages)/len(ages)
min_average=abs(min(ages)-average)
max_average=abs(max(ages)-average)
print(min_average)
print(max_average)
```

```python
ages = [19, 23, 19, 25, 21, 20, 25, 26, 25, 24] 
# Calculate average 
average = sum(ages) / len(ages) 
# Compare distances from the average 
min_diff = abs(min(ages) - average) 
max_diff = abs(max(ages) - average) 
print("abs(min - average):", min_diff) 
print("abs(max - average):", max_diff)
```

3.If list of countries has an even number of items,split into equal halves.If odd, the first half should contain one more country then second half.

```python
countries=['India','Russia','China','Finland','Sweden','Denmark','Norway']
mid=len(countries)//2
if len(countries) % 2 == 0: 
    first_half = countries[:mid] 
    second_half = countries[mid:] 
else: 
    first_half = countries[:mid + 1] 
    second_half = countries[mid + 1:] 
print("First half:", first_half) 
print("Second half:", second_half)
```

```python
countries=['India','Russia','China','Finland','Sweden','Denmark','Norway']
first_half=countries[:4]
second_half=countries[4:]
print(first_half)
print(second_half)
```
