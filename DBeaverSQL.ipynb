# Create Orders Table

```python
try:
    cursor.execute("""
        Create table Orders(
        order_id int primary key,
        order_location text,
        order_price int,
        order_quantity int)
        """)
    print('Order Table Created')
except:
    print('Order Table Already Created.')
```
