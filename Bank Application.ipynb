```python
import random
```

```python
class Bank:
    # constructor
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance
        self.account_number = ''.join(str(random.randint(1, 9)) for i in range(16))

    def print_details(self):
        print(self.account_holder)
        print(self.balance)
        print(self.account_number)

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f'Balance of Rs.{amount} deposited to A/C No. {self.account_number}.')
        else:
            print(f'Amount must be more than 0.')
            
    def withdraw(self, amount):
        if amount < self.balance:
            self.balance -= amount
            print(f'Balance of Rs.{amount} withdrawn from A/C No. {self.account_number}.')
        else:
            print(f'Amount must be less than balance amount.')
```

```python
all_accounts = []
# function used to check whether account number is available in list or not.
def findAccountByNumber(acc_number):
    for account in all_accounts:
        if account.account_number == acc_number:
            return account
    return None
```

```python
while True:
    print('1. Create New Account')
    print('2. Deposit Amount')
    print('3. Withdraw Amount')
    print('4. Print Details')
    print('5. Exit')

    choice = input('Enter a number: ')

    if choice == '1':
        name = input('Enter account holder name: ')
        balance = int(input('Enter initial balance to deposit: '))

        b = Bank(name, balance)
        all_accounts.append(b)
        print(f'Account Created with account name {name} and A/C No. {b.account_number}.')
    elif choice == '2':
        acc_number = input('Enter account number: ')
        find_acc = findAccountByNumber(acc_number)

        if find_acc:
            balance = int(input('Enter balance to deposit: '))
            find_acc.deposit(balance)
        else:
            print('Account number does not match.')
    elif choice == '3':
        acc_number = input('Enter account number: ')
        find_acc = findAccountByNumber(acc_number)

        if find_acc:
            balance = int(input('Enter balance to deposit: '))
            find_acc.withdraw(balance)
        else:
            print('Account number does not match.')
    elif choice == '4':
        acc_number = input('Enter account number: ')
        find_acc = findAccountByNumber(acc_number)

        if find_acc:
            find_acc.print_details()
        else:
            print('Account number does not match.')
    elif choice == '5':
        print('Thank you for choosing us.')
        break
    else:
        print('Invalid Input!!')
```

```python
for acc in all_accounts:
    print(acc.account_number)
    print(acc.account_holder)
    print(acc.balance)
```
